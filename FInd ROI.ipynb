**Sample code to find ROI**




Import packages

```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

Function to Mask ROI

```
def roi(img, vertices):
    #blank mask:
    mask = np.zeros_like(img)
    # fill the mask
    cv.fillPoly(mask, vertices, 255)
    # now only show the area that is the mask
    masked = cv.bitwise_and(img, mask)
    masked[:,:,1] = masked[:,:,0]
    masked[:,:,2] = masked[:,:,0]
    return masked
```

Load image

```
image_original = cv.imread('Photography.jpg')
plt.figure(1);
plt.imshow(image_original[...,::-1])
```

Generate Mask

```
roi_vertices = np.array([[10,500],[10,300],[300,200],[500,200],[800,300],[800,500],], np.int32)
roi_img = roi(image_original, [roi_vertices])
plt.figure(2)
plt.imshow(roi_img)
```

Draw a custom ROI

```
point1 = (10, 500)
point2 = (10, 300)
point3 = (300, 200)
point4 = (500, 200)
point5 = (800, 300)
point6 = (800, 500)

cv.line(image_original, point1, point2, (255,0,0), 10) 
cv.line(image_original, point2, point3, (255,0,0), 10) 
cv.line(image_original, point3, point4, (255,0,0), 10) 
cv.line(image_original, point4, point5, (255,0,0), 10) 
cv.line(image_original, point5, point6, (255,0,0), 10) 
cv.line(image_original, point6, point1, (255,0,0), 10) 
plt.figure(3);
plt.imshow(image_original[...,::-1])
```
